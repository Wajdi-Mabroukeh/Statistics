```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
# H:1, T:0
#------------------Q1----------------------
def countHeadThenTailInOneSample(n):
    count = 0;
    res = []
    trail1 = np.random.binomial(1, 0.5)
    res.append(trail1)
    for i in range(1, n):    
        trail2 = np.random.binomial(1, 0.5)
        res.append(trail2)
        if(trail1 & (not trail2)):
            count+=1
        trail1 = trail2
    return count

def getSampleValues(n, sampleMethodToRun):
    res = []
    for i in range(0, n):
        res.append(sampleMethodToRun(10))
    return res
        
ax = sns.distplot(getSampleValues(100, countHeadThenTailInOneSample))
ax.set(xlabel='Binomial Distribution', ylabel='Frequency')
plt.title("Number of sequences of \"HT\" in 10 consecutive coin tosses with 1 sample")
plt.show()
```

```python
#------------------Q2----------------------
# 1000 Sample

def ShowSamples(n, sampleMethodToRun):
    res = []
    for i in range(0, n):
        xValues = getSampleValues(100, sampleMethodToRun)
        xMean = np.mean(xValues) 
        #mean of X
        res.append(xMean)
    ax = sns.distplot(res)
    ax.set(xlabel='Binomial Distribution', ylabel='Frequency')
    

ShowSamples(1000, countHeadThenTailInOneSample)   
plt.title("Number of sequences of \"HT\" in 10 consecutive coin tosses with 1000 sample")
plt.show()
```
